```python
import matplotlib.pyplot as plt
import numpy as np

# Data for stacked bar chart: Barriers by treatment type (scale 1-6)
treatments = ['Surgery', 'Chemotherapy', 'Radiotherapy', 'Targeted Therapy', 'Immunotherapy']
cost_barrier = [4, 4, 5, 5, 6]
availability = [3, 3, 2, 1, 1]
geographic_barrier = [3, 3, 4, 5, 5]

bar_width = 0.5
indices = np.arange(len(treatments))

# Stacked Bar Chart: Barriers by Treatment Type
fig, ax = plt.subplots(figsize=(10, 6))
p1 = ax.bar(indices, cost_barrier, bar_width, label='Cost Barrier', color='#ff9999')
p2 = ax.bar(indices, availability, bar_width, bottom=cost_barrier, label='Availability Barrier', color='#66b3ff')
bottoms = np.array(cost_barrier) + np.array(availability)
p3 = ax.bar(indices, geographic_barrier, bar_width, bottom=bottoms, label='Geographic Barrier', color='#99ff99')
ax.set_xticks(indices)
ax.set_xticklabels(treatments)
ax.set_ylabel('Barrier Severity (1-6 scale)')
ax.set_title('Barriers to Accessing Cancer Treatments in Bangladesh')
ax.legend()

plt.tight_layout()
plt.show()

# Data for bar chart: Approximate access percentages
access_percentages = [70, 60, 40, 20, 10]

# Bar Chart: Access Percentage by Treatment
fig2, ax2 = plt.subplots(figsize=(8, 6))
bars = ax2.bar(treatments, access_percentages, color=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0'])
ax2.set_ylabel('Approximate Access (%)')
ax2.set_title('Access to Cancer Treatments in Bangladesh')
plt.tight_layout()
plt.show()
# Data for the charts
categories = ['Surgery', 'Chemotherapy', 'Radiotherapy', 'Targeted Therapies', 'Immunotherapies']
urban_access = [90, 85, 80, 50, 40]  # Approximate access percentages in urban areas
rural_access = [60, 55, 50, 25, 20]  # Approximate access percentages in rural areas
cost_factor = [70, 60, 50, 90, 95]  # Approximate cost impact for each treatment

# Plotting access differences between Urban and Rural Areas
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Bar chart for Urban vs Rural access
ax[0].barh(categories, urban_access, color='lightblue', label='Urban Access')
ax[0].barh(categories, rural_access, color='lightcoral', label='Rural Access')

ax[0].set_xlabel('Access Percentage (%)')
ax[0].set_title('Cancer Treatment Access: Urban vs Rural')
ax[0].legend()

# Bar chart for Cost Factor of Different Treatments
ax[1].barh(categories, cost_factor, color='orange')
ax[1].set_xlabel('Cost Impact (%)')
ax[1].set_title('Cost Impact on Cancer Treatments')

plt.tight_layout()
plt.show()
```
